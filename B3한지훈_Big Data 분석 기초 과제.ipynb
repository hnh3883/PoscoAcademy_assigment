```python
# 3~8행 선택
df.loc[3:8]
```

```python
# 5행 선택
df.iloc[5:6]
```

```python
# D > 0.5 데이터 선택
df.query('D > 0.5')
```

```python
# B > D 데이터 선택
df.query('B > D')
```

```python
# A > 0.5 그리고 D < 0.0 데이터 선택
df.query('A > 0.5 & D < 0.0')
```

```python
# A > 0.5 혹은 D < 0.0 데이터 선택
df.query('A > 0.5 | D < 0.0')
```

### 실습 5 - 데이터 결합

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                  index=[0,1,2,3])
df2 = pd.DataFrame({'E':['E4','E5','E6','E7'],
                   'F':['F4','F5','F6','F7'],
                   'G':['G4','G5','G6','G7'],
                   'H':['H4','H5','H6','H7']},
                  index=[0,1,2,3])
df3 = pd.DataFrame({'I':['I8','I9','I10','I11'],
                   'J':['J8','J9','J10','J11'],
                   'K':['K8','K9','K10','K11'],
                   'L':['L8','L9','L10','L11']},
                  index=[0,1,2,3])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# concat을 이용한 데이터 결합
result = pd.concat([df1,df2,df3])
result
```

```python
# parameter 변경
result = pd.concat([df1,df2,df3], axis = 1)
result
```

### 실습 7 - 데이터 변환
